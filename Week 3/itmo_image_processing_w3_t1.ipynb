```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 1. С помощью библиотеки OpenCV считайте изображение:

```python
file_name = "image.jpg"
img = cv2.cvtColor(cv2.imread(file_name), cv2.COLOR_BGR2GRAY)
```

### 2. Постройте гистограмму черно-белого изображения.

```python
hist_y, hist_x, _ = plt.hist(img.ravel(), bins = 256)
plt.xlabel('Интенсивность $n$')
plt.ylabel('Количество $h(n)$')
plt.show()

print(f"Максимальное значение функции h(n): {hist_y.max()}")
print(f"Интенсивность n, при которой достигается максимум h(n): {hist_x[np.where(hist_y == hist_y.max())]}")
```
