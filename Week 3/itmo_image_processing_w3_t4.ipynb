```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 1. С помощью библиотеки OpenCV считайте изображение:

```python
file_name = "image.jpg"
img = cv2.cvtColor(cv2.imread(file_name), cv2.COLOR_BGR2RGB)
```

### 2. Преобразуйте изображение в палитру RGB и примените гауссово ядро, используя метод `cv2.GaussianBlur()`, размер ядра (3, 7)

```python
img_new = cv2.GaussianBlur(img, (3, 7), 3)
plt.imshow(img_new)
```

### 3. Получите изображение с деталями.

```python
img_remaining = img - img_new
plt.imshow(img_remaining)
```
