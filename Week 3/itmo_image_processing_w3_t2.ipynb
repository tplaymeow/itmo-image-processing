```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 1. С помощью библиотеки OpenCV считайте изображение:

```python
file_name = "image.jpg"
img = cv2.cvtColor(cv2.imread(file_name), cv2.COLOR_BGR2RGB)
```

### 2. Постройте гистограмму изображения в палитре RGB. Сравните оценку плотностей распределения интенсивностей по каналам R, G и B.


```python
hist_r_y, _, _ = plt.hist(img[:, :, 0].ravel(), bins = 256, color = 'Red', alpha = 0.5, density = True)
hist_g_y, _, _ = plt.hist(img[:, :, 1].ravel(), bins = 256, color = 'Green', alpha = 0.5, density = True)
hist_b_y, _, _ = plt.hist(img[:, :, 2].ravel(), bins = 256, color = 'Blue', alpha = 0.5, density = True)
plt.xlabel('Интенсивность $n$')
plt.ylabel('Количество')
plt.show()

print(f"Максимальное значение плотности среди трех каналов: {max(hist_r_y.max(), hist_g_y.max(), hist_b_y.max())}")
```
