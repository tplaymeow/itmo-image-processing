### 1. С помощью библиотеки OpenCV считайте изображение:

```python
file_name = "image.jpg"
img = cv2.imread(file_name)
```

### 2. Определите разрешение изображения:

```python
h, w, c = img.shape
print('Width:  ', w)
print('Height: ', h)
print('Channels:', c)
```

### 3. Определите интенсивность каналов R,G,B для пикселя, находящегося на пересечении `2023` строки и `3034` столбца матричного представления изображения.

```python
B, G, R = img[2023][3034]
print('Red channel:', R)
print('Green channel:', G)
print('Blue channel:', B)
```
