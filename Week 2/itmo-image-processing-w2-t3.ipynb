```python
import cv2
print('OpenCV version:', cv2.__version__)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. С помощью библиотеки OpenCV считайте изображение:

```python
file_name = "image.jpg"
img = cv2.imread(file_name)
```

### 2. Сожмите изображение до ширины `860`, соблюдая пропорции исходного изображения. Важно: коэффициент отношения сторон при вычислениях не округляйте. Если высота не окажется целой, округлите значение до целого в меньшую сторону. Используйте `cv2.resize()` с параметром `interpolation = cv2.INTER_AREA`:

```python
h, w, c = img.shape
aspect_ratio = w / h
resized_w = 860
resized_h = int(resized_w / aspect_ratio)
resized_dim = (resized_w, resized_h)
resized_img = cv2.resize(img, resized_dim, interpolation = cv2.INTER_AREA)

print(f'Aspect ratio: {aspect_ratio}')
print(f'Resized height: {resized_h}')
plt.imshow(resized_img)
plt.show()
```

### 3. Сохраните сжатое изображение в палитре RGB:

```python
output_file_name = "output_file.jpg"
cv2.imwrite(output_file_name, resized_img)
```
